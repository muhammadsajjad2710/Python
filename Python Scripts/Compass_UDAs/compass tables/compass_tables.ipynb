```python
import mysql.connector
import pandas as pd
import numpy as np

# Database credentials and configuration
username = 'root'
password = ''
database_name = 'new'
host = 'localhost'

# Establish connection to MySQL
try:
    conn = mysql.connector.connect(
        user=username,
        password=password,
        host=host,
        database=database_name
    )
    cursor = conn.cursor()
    print("Connected to MySQL database:", conn.database)
except mysql.connector.Error as err:
    print("Error connecting to MySQL:", err)

# Read the dataset
dataset_path = 'contract_id.csv'  # Replace with the path to your dataset
df = pd.read_csv(dataset_path)

# Ensure the DataFrame columns match the table structure
# Assuming dataset columns are exactly as described
df.columns = ['id', 'Contract_ID', 'Commissioner', 'ContractType', 'Start_Date', 'End_Date', 'Status', 'Contract_UDAs']

# Handle NaN values if any
df = df.replace({np.nan: None})

# Insert data into the MySQL table
for index, row in df.iterrows():
    sql = """
    INSERT INTO contract_id (id, Contract_ID, Commissioner, ContractType, Start_Date, End_Date, Status, Contract_UDAs)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(sql, tuple(row))

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()
print("Data inserted successfully.")
```

```python
import mysql.connector
import pandas as pd
import numpy as np

# Replace these values with your MySQL credentials and database information
username = 'root'
password = ''
database_name = 'new'
host = 'localhost'  # e.g., 'localhost'

# Establish connection to MySQL
try:
    conn = mysql.connector.connect(
        user=username,
        password=password,
        host=host,
        database=database_name
    )
    print("Connected to MySQL database:", conn.database)
except mysql.connector.Error as err:
    print("Error connecting to MySQL:", err)

# Prepare SQL insert statement
insert_statement = """
INSERT INTO data_report
(id, CRN_IN, ContractID, PerformerID, Errors, PatientSurname, PatientForename,
 DateOfBirth, TreatmentStartDate, TreatmentEndDate, PatientCharge, Period,
 Units, ContraPeriod_Units_PCharge, NHSNumber, ProcessingDate, Status,
 FormType, CreatedAmendedDeletedby, CreationAmendDeletionDate)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

try:
    with conn.cursor() as cursor:
        # Iterate over DataFrame rows
        for index, row in df.iterrows():
            # Handle NaN values by converting them to None
            row = row.where(pd.notnull(row), None)
            
            # Extract values from DataFrame row and perform necessary conversions
            values = (
                row['id'],
                row['CRN_IN'],
                row['ContractID'],
                row['PerformerID'],
                row['Errors'],
                row['PatientSurname'],
                row['PatientForename'],
                row['DateOfBirth'],
                row['TreatmentStartDate'],
                row['TreatmentEndDate'],
                row['PatientCharge'],
                row['Period'],
                row['Units'],
                row['ContraPeriod_Units_PCharge'],
                row['NHSNumber'],
                row['ProcessingDate'],
                row['Status'],
                row['FormType'],
                row['CreatedAmendedDeletedby'],
                row['CreationAmendDeletionDate']
            )
            
            try:
                # Execute SQL insert statement
                cursor.execute(insert_statement, values)
                conn.commit()  # Commit each insertion immediately
            except mysql.connector.IntegrityError as e:
                # Handle duplicate primary key error
                print(f"Skipped duplicate record with id {row['id']}")
                conn.rollback()  # Rollback the transaction
            
    print("Data insertion completed")

except Exception as e:
    print(f"Error occurred: {e}")

finally:
    # Close connection
    conn.close()
```
