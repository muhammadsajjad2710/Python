```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_path = r'C:\Users\Awan\Downloads\download.jfif'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python

# Histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot original and equalized images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

# histogram matching
# Load the reference image
reference_image_path = 'newcar.jpeg'
reference_image = cv2.imread(reference_image_path, cv2.IMREAD_GRAYSCALE)

# Perform histogram matching
matched_image = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8)).apply(image)

# Plot original and matched images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(matched_image, cmap='gray')
plt.title('Histogram Matched Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
