```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_path = r'C:\Users\Awan\Downloads\download.jfif'

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

def negative_transformation(image):
    return 255 - image

def logarithmic_transformation(image):
    c = 255 / np.log(1 + np.max(image))
    log_image = c * (np.log(image + 1))
    return np.array(log_image, dtype=np.uint8)

def power_law_transformation(image, gamma):
    gamma_corrected = np.array(255 * (image / 255) ** gamma, dtype=np.uint8)
    return gamma_corrected

negative_image = negative_transformation(image)
logarithmic_image = logarithmic_transformation(image)
gamma_image = power_law_transformation(image, gamma=0.5)

images = [image, negative_image, logarithmic_image, gamma_image]
titles = ['Original Image', 'Negative Image', 'Logarithmic Image', 'Gamma Image (γ=0.5)']

plt.figure(figsize=(10, 8))
for i in range(len(images)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
