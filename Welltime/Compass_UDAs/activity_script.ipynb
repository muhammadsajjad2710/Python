```python
import os
import pandas as pd
import pyodbc
import mysql.connector

# Excel file path
excel_file_path = r'C:\Users\Awan\Downloads\DCS126_RGHAURI.xlsx'

# SQL Server connection details with Windows authentication
server_name = 'IMRANKHAN\\SQLEXPRESS'  # Note the double backslash
database_name = 'postgres'

# Establish a connection to the SQL Server database with integrated security
conn_str = f'DRIVER={{SQL Server}};SERVER={server_name};DATABASE={database_name};Trusted_Connection=yes;'
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

try:
    # Read Excel file into a pandas DataFrame starting from row 9
    df = pd.read_excel(excel_file_path, skiprows=7)
    # Convert all columns to strings (varchar(50))
    df = df.applymap(str)
    
    for index, row in df.iterrows():
        # Check if a row with the same key already exists
        cursor.execute('''
            SELECT 1 
            FROM data_report 
            WHERE 
                CRN_IN = ? AND ContractID = ? AND PerformerID = ? AND 
                PatientSurname = ? AND PatientForename = ? AND DateOfBirth = ?
        ''',
        (row['CRN/IN'], row['Contract ID'], row['Performer ID'], row['Patient Surname'],
         row['Patient Forename'], row['Date of Birth']))
        
        existing_row = cursor.fetchone()

        if existing_row:
            # Update the existing row
            cursor.execute('''
                UPDATE data_report 
                SET 
                    Errors = ?, PatientCharge = ?, Period = ?, Units = ?, 
                    ContraPeriod_Units_PCharge = ?, [NHS Number] = ?, 
                    ProcessingDate = ?, Status = ?, FormType = ?, 
                    Created_Amended_Deletedby = ?, Creation_Amend_DeletioDate = ?
                WHERE 
                    CRN_IN = ? AND ContractID = ? AND PerformerID = ? AND 
                    PatientSurname = ? AND PatientForename = ? AND DateOfBirth = ?
            ''',
            (row['Errors'], row['Patient Charge'], row['Period'], row['Units'],
             row['Contra Period, Units, P. Charge'], row['NHS Number'],
             row['Processing Date'], row['Status'], row['Form Type'],
             row['Created/Amended/Deleted by'], row['Creation/Amend/Deletion Date'],
             row['CRN/IN'], row['Contract ID'], row['Performer ID'],
             row['Patient Surname'], row['Patient Forename'], row['Date of Birth']))
        else:
            # Insert a new row if no matching row is found
            cursor.execute('''
                INSERT INTO data_report 
                (CRN_IN, ContractID, PerformerID, Errors, PatientSurname, PatientForename, DateOfBirth, 
                TreatmentStartDate, TreatmentEndDate, PatientCharge, Period, Units, ContraPeriod_Units_PCharge, 
                [NHS Number], ProcessingDate, Status, FormType, Created_Amended_Deletedby, Creation_Amend_DeletioDate)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            ''', 
            (row['CRN/IN'], row['Contract ID'], row['Performer ID'], row['Errors'], row['Patient Surname'],
             row['Patient Forename'], row['Date of Birth'], row['Treatment Start Date'], row['Treatment End Date'],
             row['Patient Charge'], row['Period'], row['Units'], row['Contra Period, Units, P. Charge'],
             row['NHS Number'], row['Processing Date'], row['Status'], row['Form Type'],
             row['Created/Amended/Deleted by'], row['Creation/Amend/Deletion Date']))

    # Commit the changes
    conn.commit()
  

    # Delete the Excel file after successful data insertion or update
    os.remove(excel_file_path)
   

except Exception as e:
    print(f"Error: {e}")

finally:
    # Close the cursor and connection
    cursor.close()
    conn.close()
```
